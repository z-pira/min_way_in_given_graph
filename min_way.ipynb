```python
v = 8
INF = sys.maxsize

def min_way(u, destination, visited, graph):

    if (u == destination):
        return 0

    visited[u] = 1
    ans = INF

    for i in range(v):
        if (graph[u][i] != INF and not visited[i]):
            curr = min_way(i, destination, visited, graph)

            if (curr < INF):
                ans = min(ans, graph[u][i] + curr)

    visited[u] = 0

    return ans

if __name__=="__main__":

        graph = [[INF for j in range(v)]
                for i in range(v)]

        visited = [0 for i in range(v)]

        graph[0][1] = 624
        graph[0][2] = 1038
        graph[1][0] = 624
        graph[1][3] = 895
        graph[1][4] = 924
        graph[2][0] = 1038
        graph[2][3] = 481
        graph[2][4] = 1238
        graph[2][7] = 765
        graph[3][1] = 895
        graph[3][2] = 481
        graph[3][5] = 440
        graph[4][1] = 924
        graph[4][2] = 1238
        graph[4][5] = 1306
        graph[5][3] = 440
        graph[5][4] = 1306
        graph[5][7] = 1008
        graph[6][7] = 1169
        graph[7][2] = 765
        graph[7][5] = 1008
        graph[7][6] = 1169

        s = 0
        t = 2

        visited[s] = 0
        print(min_way(s, t, visited, graph))
```
